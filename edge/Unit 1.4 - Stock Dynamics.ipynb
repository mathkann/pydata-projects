```python
% matplotlib inline
import pandas as pd
import numpy as np
import os
pd.options.display.mpl_style = 'default'
import matplotlib.pyplot as plt
```

## Loading in the data

```python
data_folder = '..' + os.sep + 'data' + os.sep

# loading the files
# we want to load the first column as dates, so we use the 'parse_dates' and 'infer_datetime_format' args to help
IBM = pd.read_csv(data_folder + 'IBMStock.csv', parse_dates=[0], infer_datetime_format=True)
GE = pd.read_csv(data_folder + 'GEStock.csv', parse_dates=[0], infer_datetime_format=True)
Boeing = pd.read_csv(data_folder + 'BoeingStock.csv', parse_dates=[0], infer_datetime_format=True)
ProcterGamble = pd.read_csv(data_folder + 'ProcterGambleStock.csv', parse_dates=[0], infer_datetime_format=True)
CocaCola = pd.read_csv(data_folder + 'CocaColaStock.csv', parse_dates=[0], infer_datetime_format=True)
```

## Summary Statistics

```python
# What is the mean stock price of IBM over this time period?
print 'Mean Stock Price of IBM: ', IBM.StockPrice.mean()

# What is the minimum stock price of General Electric (GE) over this time period?
print 'Minimum stock price of General Electric (GE): ', GE.StockPrice.min()

# What is the maximum stock price of Coca-Cola over this time period?
print 'Maximum stock price of Coca-Cola: ', CocaCola.StockPrice.max()

# What is the median stock price of Boeing over this time period?
print 'Median stock price of Boeing: ', Boeing.StockPrice.median()

# What is the standard deviation of the stock price of Procter & Gamble over this time period?
print 'Standard Deviation of the stock price of Procter & Gamble: ', ProcterGamble.StockPrice.std()
```

## Visualizing Stock Dynamics

```python
# Line plot of Stock Price of Coca-Cola
CocaCola.plot(kind='line', x='Date', y='StockPrice')
```

```python

# Create a combined data frame
combined_df = CocaCola.merge(ProcterGamble, on='Date', suffixes=('_CocaCola', '_ProcterGamble'))

combined_df.head()
# Multiple line plots
combined_df.plot(x='Date')
```

### Combine & Comapre all datasets

```python
combined_df = combined_df.merge(IBM, on='Date')
combined_df = combined_df.merge(GE, on='Date', suffixes=('_IBM', '_GE'))
combined_df = combined_df.merge(Boeing, on='Date')

# Rename  'StockPrice' column for Boeing
combined_df.rename(columns={'StockPrice':'StockPrice_Boeing'}, inplace=True)
```

```python
combined_df.head()
```

### Which stock fell the most right after the technology bubble burst in March 2000?

```python
# Plotting the different Stock Prices from 1995 to 2005

f = plt.figure()
plt.title('Stock Prices Comparison', color='black')
combined_df.iloc[300:432].plot(x='Date', ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```
