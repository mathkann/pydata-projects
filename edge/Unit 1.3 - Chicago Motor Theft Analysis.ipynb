## Dates in Pandas

### Convert columns to Date

```python
# convert the 'Date' column in our dataset to Date

date_convert = pd.to_datetime(mvt['Date'])
date_convert.head()
```

```python
date_convert.describe()
```

```python
# Extract month , day of week and add to our data frame
mvt['Month'] = pd.DatetimeIndex(date_convert).month
mvt['Weekday'] = pd.DatetimeIndex(date_convert).dayofweek
mvt['Date'] = date_convert
```

```python
# In which month did the fewest motor vehicle thefts occur?
mvt.groupby('Month').count().sort('Arrest')
```

```python
# On which weekday did the most motor vehicle thefts occur?
mvt.groupby('Weekday').count().sort('Arrest', ascending=False)
```

```python
# Which month has the largest number of motor vehicle thefts for which an arrest was made?
(mvt[mvt['Arrest'] == True]).groupby('Month').count().sort('Arrest', ascending=False)
```
