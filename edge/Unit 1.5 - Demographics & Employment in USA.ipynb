## Integrating Metropolitan & Country Data

```python
# Read in the Metro Area Code and the Country Codes
MetroAreaMap  = pd.read_csv(data_folder + 'MetroAreaCodes.csv')
CountryMap = pd.read_csv(data_folder + 'CountryCodes.csv')
```

```python
MetroAreaMap.describe()
```

```python
CountryMap.describe()
```

### Merge CPS data frame with the Metropolitan Area data

```python
# do a left-outer join ...that is we retain all rows of CPS irrespective of a match in the MetroAreaMap data frame
# this ensures, we do not delete rows that have a NA value in the 'MetroAreaCode' 

CPS_merged = CPS.merge(MetroAreaMap, how='left', left_on='MetroAreaCode', right_on='Code')
```

```python
CPS_merged.dtypes
```

### Which of the following metropolitan areas has the largest number of interviewees?

## Integrating Country of Birth Data

```python
CPS_merged.drop('Code', axis=1, inplace=True)
```

```python
CPS_merged = CPS_merged.merge(CountryMap, how='left', left_on='CountryOfBirthCode', right_on='Code')
```

```python
CPS_merged.dtypes
```

### How many interviewees have a missing value for the new country of birth variable?

```python
sum(CPS_merged.Country.isnull())
```

### Among all interviewees born outside of North America, which country was the most common place of birth?

```python
print_full(CPS_merged.groupby('Country').agg(['count']))
```

### What proportion of the interviewees from the "New York-Northern New Jersey-Long Island, NY-NJ-PA" metropolitan area have a country of birth that is not the United States? 

```python
CPS_merged[CPS_merged.MetroArea == 'New York-Northern New Jersey-Long Island, NY-NJ-PA' and CPS_merged.Country != 'United States']
```
